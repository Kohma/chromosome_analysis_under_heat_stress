# Accessibility modeling under heat stress

```python
import numpy as np
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
from numpy.random import *
```

```python
#　definistion function
def generate_points(num):
    Points = []
    for i in range(num):
        # dotの発生
        x1 = 2*(rand()-0.5)
        y1 = 2*(rand()-0.5)
        z1 = 2*(rand()-0.5)
        while x1*x1+y1*y1+z1*z1 > 1 or  x1*x1+y1*y1+z1*z1 < 0.7:
            x1 = 2*(rand()-0.5)
            y1 = 2*(rand()-0.5)
            z1 = 2*(rand()-0.5)
        point = ([x1,y1,z1,0])
        Points.append(point)
    return Points

def Random_walk(points,mobility):
    for i in range(len(points)):
        #determination of the smallness mobility
        x1,y1,z1 = points[i][0],points[i][1],points[i][2]
        dx1 = 2*(rand()-0.5)*mobility
        dy1 = 2*(rand()-0.5)*mobility
        dz1 = 2*(rand()-0.5)*mobility
        while dx1*dx1+dy1*dy1+dz1*dz1 > 0.01*mobility*mobility or (x1+dx1)*(x1+dx1)+(y1+dy1)*(y1*dy1)+(z1+dz1)*(z1+dz1) > 1\
        or (x1+dx1)*(x1+dx1)+(y1+dy1)*(y1+dy1)+(z1+dz1)*(z1+dz1) < 0.7:
            dx1 = 2*(rand()-0.5)*mobility
            dy1 = 2*(rand()-0.5)*mobility
            dz1 = 2*(rand()-0.5)*mobility
        points[i] = np.array(points[i])+np.array([dx1,dy1,dz1,0])
    return points

def activation_nuc_point(Nuc_points, Act_points):
    for i in range(len(Nuc_points)):
        if Nuc_points[i][3] == 0:
            for j in range(len(Act_points)):
                magnif = 100
                x1,y1,z1 = Nuc_points[i][0]*magnif,Nuc_points[i][1]*magnif,Nuc_points[i][2]*magnif
                x2,y2,z2 = Act_points[j][0]*magnif,Act_points[j][1]*magnif,Act_points[j][2]*magnif
                dist = (x1-x2)**2+(y1-y2)**2+(z1-z2)**2
                if dist < 10:
                    Nuc_points[i][3] = 1
    return Nuc_points
            
def plot_figure(Nuc_points,t):
    Activated_points = []
    Not_Activated = []
    for i in range(len(Nuc_points)):
        if Nuc_points[i][3] == 0:
            not_activated = ([Nuc_points[i][0],Nuc_points[i][1],Nuc_points[i][2]])
            Not_Activated.append(not_activated)
        else:
            activated = ([Nuc_points[i][0],Nuc_points[i][1],Nuc_points[i][2]])
            Activated_points.append(activated)
    Act = np.array(Activated_points)
    NoAct = np.array(Not_Activated)
    X1,Y1,Z1 = Act[:,0],Act[:,1],Act[:,2]
    X2,Y2,Z2 = NoAct[:,0],NoAct[:,1],NoAct[:,2]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter3D(X1, Y1, Z1, s=2, color="red")
    ax.scatter3D(X2, Y2, Z2, s=2,color = "blue")
    os.chdir(r"C:\Users\***\modeling_result_graph\control") #chose working directory
    name = "time_"+ str(t) +".png"
    plt.savefig(name)
    #plt.scatter(X1,Y1,s=1,color='green')
    #plt.scatter(X2,Y2,s=1,color='red')
    return Act, NoAct
```

```python
## modeling
control_mobility = 1
under_heat_mobility = 4
number_of_loop = 1000

#modeling control 
flag = 0
if flag == 0:
    mobility = control_mobility
else:
    mobility = under_heat_mobility

#　generate point on the nucleus
Number_of_Nuc_point = 1000
Nuc_points = generate_points(Number_of_Nuc_point)

# generate nucleus factor activator
Number_of_Nuc_Activator = 1000
Nuc_Activator = generate_points(Number_of_Nuc_Activator)

#print(len(Nuc_points),len(Nuc_Activator))

# Random walk of the nucleus factor activator
result = open("20181014_Simulation_Test_controle_1000_save1.txt","w")
for t in range(number_of_loop):
    Nuc_points = Random_walk(Nuc_points,mobility*0.01)
    Nuc_Activator = Random_walk(Nuc_Activator,mobility*0.01)
    Nuc_points = activation_nuc_point(Nuc_points, Nuc_Activator)
    Act, NoAct = plot_figure(Nuc_points,t)
    print(t, Act.shape[0], NoAct.shape[0])
    result.write("%d\t%d\t%d\n"%(t,Act.shape[0], NoAct.shape[0]))
result.close()
```
