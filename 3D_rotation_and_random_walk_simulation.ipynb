# 3D rotation and random walk simulation

```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import *
os.chdir(r"C:\Users\___\Desktop") # choose working directory

result4 = open("result4.txt","w")
cycle= 100
ANG1 = []
ANG2 = []
# 3D rotation
for t in range(1000):
    Ct1 = 0
    Ct2 = 0
    Ct3 = 0

    for i in range(cycle):
        # generate random dot1
        x1 = 2*(rand()-0.5)
        y1 = 2*(rand()-0.5)
        z1 = 2*(rand()-0.5)
        while x1*x1+y1*y1+z1*z1 > 1 and  x1*x1+y1*y1+z1*z1 < 0.7:
            x1 = 2*(rand()-0.5)
            y1 = 2*(rand()-0.5)
            z1 = 2*(rand()-0.5)
        A1 = ([x1,y1,z1])

        # generate random dot2
        x2 = 2*(rand()-0.5)
        y2 = 2*(rand()-0.5)
        z2 = 2*(rand()-0.5)
        while x2*x2+y2*y2+z2*z2 > 1 and  x2*x2+y2*y2+z2*z2 < 0.7:
            x2 = 2*(rand()-0.5)
            y2 = 2*(rand()-0.5)
            z2 = 2*(rand()-0.5)
        A2 = ([x2,y2,z2])

        # generate rotation axis
        nx = 2*(rand()-0.5)
        ny = 2*(rand()-0.5)
        nz = 2*(rand()-0.5)
        while nx*nx+ny*ny+nz*nz >1:
            nx = 2*(rand()-0.5)
            ny = 2*(rand()-0.5)
            nz = 2*(rand()-0.5)
        N = np.sqrt(nx*nx+ny*ny+nz*nz)
        nx=nx/N
        ny=ny/N
        nz=nz/N

        # generatte rotation angle
        theta = 2*(rand()-0.5)*np.pi/100

        # define rotation matrix
        Rn = np.array([[np.cos(theta)+ nx*nx*(1-np.cos(theta)), nx*ny*(1-np.cos(theta))-nz*np.sin(theta),\
                   nx*nz*(1-np.cos(theta))+ny*np.sin(theta)],\
                 [ny*nx*(1-np.cos(theta))+nz*np.sin(theta), np.cos(theta)+ny*ny*(1-np.cos(theta)),\
                  ny*nz*(1-np.cos(theta))-nx*np.sin(theta)],\
                 [nz*nx*(1-np.cos(theta))-ny*np.sin(theta), nz*ny*(1-np.cos(theta))+nx*np.sin(theta),\
                  np.cos(theta)+nz*nz*(1-np.cos(theta))]])
        B1 = np.dot(Rn,A1)
        B2 = np.dot(Rn,A2)

        ang1 = np.arctan((B1[1]-A1[1])/(B1[0]-A1[0]))
        ang2 = np.arctan((B2[1]-A2[1])/(B2[0]-A2[0]))
        d1=(A2[0]-A1[0])*(A2[0]-A1[0])+(A2[1]-A1[1])*(A2[1]-A1[1])
        d2=(B2[0]-B1[0])*(B2[0]-B1[0])+(B2[1]-B1[1])*(B2[1]-B1[1])
        if ang2 < ang1 +0.1 and ang2 > ang1 -0.1:
            Ct1 = Ct1+1
        if 0.6 < np.absolute(ang1) < 1.4 and 0.6 < np.absolute(ang2) <1.4:
            Ct2 = Ct2+1
        if ang1*ang1+ang2*ang2 < 0.4:
            Ct3 = Ct3+1
    CT1 = Ct1/cycle
    CT2 = Ct2/cycle
    CT3 = Ct3/cycle
    
    result4.write("0\t%f\t%f\t%f\n" %(CT1,CT2,CT3))

# Random walk
for t in range(1000):
    Ct1 = 0
    Ct2 = 0
    Ct3 = 0
    for i in range(cycle):
        # generate random dot1
        x1 = 2*(rand()-0.5)
        y1 = 2*(rand()-0.5)
        z1 = 2*(rand()-0.5)
        while x1*x1+y1*y1+z1*z1 > 1 and  x1*x1+y1*y1+z1*z1 < 0.7:
            x1 = 2*(rand()-0.5)
            y1 = 2*(rand()-0.5)
            z1 = 2*(rand()-0.5)
        A1 = np.array([x1,y1,z1])

        # generate random dot2
        x2 = 2*(rand()-0.5)
        y2 = 2*(rand()-0.5)
        z2 = 2*(rand()-0.5)
        while x2*x2+y2*y2+z2*z2 > 1 and  x2*x2+y2*y2+z2*z2 < 0.7:
            x2 = 2*(rand()-0.5)
            y2 = 2*(rand()-0.5)
            z2 = 2*(rand()-0.5)
        A2 = np.array([x2,y2,z2])

        # minute movement amount
        dx1 = 2*(rand()-0.5)*0.01
        dy1 = 2*(rand()-0.5)*0.01
        dz1 = 2*(rand()-0.5)*0.01
        while dx1*dx1+dy1*dy1+dz1*dz1 > 0.01 and (x1+dx1)*(x1+dx1)+(y1+dy1)*(y1*dy1)+(z1+dz1)*(z1+dz1) > 1\
        and (x1+dx1)*(x1+dx1)+(y1+dy1)*(y1*dy1)+(z1+dz1)*(z1+dz1) < 0.7:
            dx1 = 2*(rand()-0.5)*0.01
            dy1 = 2*(rand()-0.5)*0.01
            dz1 = 2*(rand()-0.5)*0.01
        dA1 = np.array([dx1,dy1,dz1])

        dx2 = 2*(rand()-0.5)*0.01
        dy2 = 2*(rand()-0.5)*0.01
        dz2 = 2*(rand()-0.5)*0.01
        while dx2*dx2+dy2*dy2+dz2*dz2 > 0.01 and (x2+dx2)*(x2+dx2)+(y2+dy2)*(y2*dy2)+(z2+dz2)*(z2+dz2) > 1\
        and (x2+dx2)*(x2+dx2)+(y2+dy2)*(y2*dy2)+(z2+dz2)*(z2+dz2) < 0.7:
            dx2 = 2*(rand()-0.5)*0.01
            dy2 = 2*(rand()-0.5)*0.01
            dz2 = 2*(rand()-0.5)*0.01
        dA2 = np.array([dx2,dy2,dz2])

        B1=A1+dA1
        B2=A2+dA2

        ang1 = np.arctan(dy1/dx1)
        ang2 = np.arctan(dy2/dx2)    
        
        if ang2 < ang1 +0.1 and ang2 > ang1 -0.1:
            Ct1 = Ct1+1
        if 0.6 < np.absolute(ang1) < 1.4 and 0.6 < np.absolute(ang2) <1.4:
            Ct2 = Ct2+1
        if ang1*ang1+ang2*ang2 < 0.4:
            Ct3 = Ct3+1
    CT1 = Ct1/cycle
    CT2 = Ct2/cycle
    CT3 = Ct3/cycle
    
    result4.write("1\t%f\t%f\t%f\n" %(CT1,CT2,CT3))
result4.close()    

    #print(ang1, ang2)
    #dot = dot.append([ang1, ang2])
```
